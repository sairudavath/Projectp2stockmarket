# ACTUAL CODE FOR SPARK QUERIES

```python



from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.functions import col,sum

from pyspark.sql.functions import desc
from pyspark.sql.functions import min
from pyspark.sql.functions import max

# File location and type
file_location = "/FileStore/tables/stock_data.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "True"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)

def display_menu():
    
    #print("\\\\\\\\Stock Market Data Analysis Menu\\\\\\\\\\\\\\")
    print('''
            "\\\\\\\\\\Stock Market Data Analysis Menu\\\\\\\\\\\\\\"
            
            1.  Query to show the First 5 Rows from the dataset
            2.  Query to find the total number of Rows and Columns in the dataset
            3.  Query to find the total number of stocks in the dataset
            4.  Query to find the average open price of each stock
            5.  Query to find the highest and lowest current price of each stock
            6.  Query to find the average high price of each stock
            7.  Calculate the total volume traded for each "Stocks" and sort the result in descending order
            8.  Query to find the average volume of each stock traded
            9.  Query to find the date with the highest and lowest trading volume for given stock(RELIANCE)
            10. Query to find the highest and lowest opening price of each stock
            11. Query to find the Minimum and Maximum Median of opening price for each stock
            12. Query to find the highest and lowest closing price for each stock
            13. Calculate the average Open price for each stock and order the result in descending order
            14. Calculate the Standard Deviation of "High" and "Low" for each row
            15. Query to find the Average of High and Median of Low 
            16. Query to find the Median of Current prize of each stock
            17. Filter the DataFrame to show only the rows where "Stocks" is "Reliance"
            18. Query to calculate the total trading volume for each stock
            19. Get the top 5 stocks by Maximum volume
            20. Get the lowest 5 stocks by Maximum volume
            21. EXIT
                    ''')
    
    choice = input("Enter your choice (1-11): ")
    return int(choice)

    
while True:
    
    # Ask the user for their username and password to validate their access
    admins = {
                "u": "p"
             }
    
    users = {
                "s": "r",
                "jai": "surya",
                "user3": "password3"
            }


    
    role = input("Enter your role (Admin/User): ")
    if role.lower() == "admin":
        admin_username = input("Enter your username: ")
        admin_password = input("Enter your password: ")
        if admin_username in admins and admins[admin_username] == admin_password:
            print("Access granted! You are logged in as an admin.")
            
                # its working 
            
        else:
            print("Access denied! Invalid username or password.")
            continue
            
            
    elif role.lower() == "user":
        user_username = input("Enter your username: ")
        user_password = input("Enter your password: ")
        if user_username in users and users[user_username] == user_password:
            print("Access granted! You are logged in as a user.")
            display_menu()
        else:
            print("Access denied! Invalid username or password.")
            continue
    else:
        print("Invalid role entered. Please try again.")
        continue

        
        
    choice = display_menu()

    if choice == 1:
        # 1. Query to show the First 5 Rows from the dataset
        print(df.show(5))
        
    elif choice == 2:
        # 2. Query to find the total number of Rows and Columns in the dataset
        print("Number of rows:", df.count())
        print("Number of columns:", len(df.columns))
        
    elif choice == 3:
        # 3. Query to find the total number of STOCKS in the dataset
        distinct_stocks = df.select("Stock").distinct().collect()
        total_stocks = len(distinct_stocks)
        print("Total number of stocks:", total_stocks)
        print("Different stocks in the dataset are:")
        
        for stock in distinct_stocks:
            print(stock[0])
        
        
    elif choice == 4:
        # 4. Query to find the average OPEN price of each stock
        avg_open_price = df.groupBy("Stock").agg(round(avg("Open"), 2).alias("Average Open Price"))
        avg_open_price.show()
        
    
    elif choice == 5:
        # 5. Query to find the highest and lowest CURRENT price of each stock
        highest_current_price = df.groupBy("Stock").agg(max("Current").alias("Highest Current Price"))
        lowest_current_price = df.groupBy("Stock").agg(min("Current").alias("Lowest Current Price"))
        current_price = highest_current_price.join(lowest_current_price, "Stock")
        current_price.show()
        
    elif choice == 6:
        # 6. Query to find the average HIGH price of each stock
        avg_high_price = df.groupBy("Stock").agg(round(avg("High"), 2).alias("Average High Price"))
        avg_high_price.show()
        
    
    elif choice == 7:
        # 7. Calculate the total VOLUME traded for each STOCKS and sort the result in descending order
        total_volume_df = df.groupBy("Stock").agg(sum("Volume").alias("Total_Volume")).orderBy("Total_Volume", ascending=False)
        total_volume_df.show()
        
    elif choice == 8:
        # 8. Query to find the average volume of each stock traded
        avg_volume = df.groupBy("Stock").agg(round(avg("Volume"), 2).alias("Average Volume Traded"))
        avg_volume.show()
        
    elif choice == 9:
        # 9. Query to find the date with the highest and lowest trading volume for given stock(RELIANCE)
        reliance_df = df.filter(df["Stock"] == "RELIANCE")
        if reliance_df.count() > 0:
            highest_trading_date = reliance_df.orderBy(desc("Volume")).select("Log_datetime").head()[0]
            lowest_trading_date = reliance_df.orderBy("Volume").select("Log_datetime").head()[0]
            print("Date with the highest trading volume for Reliance stock:", highest_trading_date)
            print("Date with the lowest trading volume for Reliance stock:", lowest_trading_date)
        else:
            print("No data available for Reliance stock.")

        
    elif choice == 10:
        # 10. Query to find the highest and lowest opening price of each stock
        highest_open_price = df.groupBy("Stock").agg(max("Open").alias("Highest Open Price"))
        lowest_open_price = df.groupBy("Stock").agg(min("Open").alias("Lowest Open Price"))
        open_price = highest_open_price.join(lowest_open_price, "Stock")
        open_price.show()
        
    elif choice == 11:
        # 11. Query to find the date with the highest and lowest opening price for each stock
        highest_opening_date = df.orderBy(desc("Open")).select("Log_datetime").first()[0]
        lowest_opening_date = df.orderBy("Open").select("Log_datetime").first()[0]
        highest_opening_date_per_stock = df.orderBy(desc("Open")).groupBy("Stock").agg(first("Log_datetime").alias("Highest Opening Date"))
        lowest_opening_date_per_stock = df.orderBy("Open").groupBy("Stock").agg(first("Log_datetime").alias("Lowest Opening Date"))
        opening_dates = highest_opening_date_per_stock.join(lowest_opening_date_per_stock, "Stock")
        opening_dates.show()
        
    elif choice == 12:
        # 12. Query to find the date with the highest and lowest closing price for each stock
        highest_closing_date = df.orderBy(desc("Current")).select("Log_datetime").first()[0]
        lowest_closing_date = df.orderBy("Current").select("Log_datetime").first()[0]
        highest_closing_date_per_stock = df.orderBy(desc("Current")).groupBy("Stock").agg(first("Log_datetime").alias("Highest Closing Date"))
        lowest_closing_date_per_stock = df.orderBy("Current").groupBy("Stock").agg(first("Log_datetime").alias("Lowest Closing Date"))
        closing_dates = highest_closing_date_per_stock.join(lowest_closing_date_per_stock, "Stock")
        closing_dates.show()
        
    elif choice == 13:
        # 13. Calculate the average Open price for each stock and order the result in descending order
        avg_open_df = df.groupBy("Stock").agg(avg("Open").alias("Average_Open")).orderBy(col("Average_Open").desc())
        avg_open_df.show()    
    
    elif choice == 14:
        # 14. Calculate the difference between "High" and "Low" for each row
        df_with_diff = df.withColumn("Difference", col("High") - col("Low"))
        df_with_diff.show()
        
    elif choice == 15:
        # 15. Filter the DataFrame to show only the rows where the difference between High and Low is above a threshold
        threshold = 100.0
        filtered_diff_df = df.filter((col("High") - col("Low")) > threshold)
        filtered_diff_df.show()
        
    elif choice == 16:
        # 16. Filter the DataFrame to show only the rows where "Current" is above a certain threshold
        threshold = 2500.0
        filtered_current_df = df.filter(df["Current"] > threshold)
        filtered_current_df.show()
        
    elif choice == 17:
        # 17. Filter the DataFrame to show only the rows where "Stocks" is "Reliance"
        reliance_df = df.filter(df["Stock"] == "RELIANCE")
        reliance_df.show()
        
    elif choice == 18:
        # 18. Query to calculate the total trading volume for each stock and each date
        total_volume_per_stock_date = df.groupBy("Stock", "Log_Datetime").agg(sum("Volume").alias("Total Volume"))
        total_volume_per_stock_date.show()
    
    elif choice == 19:
        # 19. Get the top 5 max stocks by volume
        top_5_stocks_by_volume = df.orderBy(col("Volume").desc()).limit(5)
        # Select the "Stocks" column along with other columns
        top_5_stocks_with_names = top_5_stocks_by_volume.select("Stock", "Volume")
        top_5_stocks_with_names.show()

    elif choice == 20:
        # 20. Get the top 5 min stocks by volume
        top_stocks_by_min_volume = df.withColumn("Volume", col("Volume").cast("double")) \
        .groupBy("Stock").min("Volume").orderBy(col("min(Volume)")).limit(5)
        top_stocks_by_min_volume.show()

           
    elif choice == 21:
        print("Exiting The Program")
        break
    else:
        print("Invalid choice. Please try again.")







```
